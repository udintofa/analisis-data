```python
import pandas as pd
from sqlalchemy import create_engine

df = pd.read_csv("../data/sales.csv")

df.columns = [
    "order_id", "order_date", "item_name", "item_type",
    "item_price", "quantity", "transaction_amount",
    "transaction_type", "received_by", "time_of_sale"
]

df.info()
```

```python
df["order_date"] = (
    df["order_date"]
    .astype(str)
    .str.strip()
    .str.replace(".", "/", regex=False)
    .str.replace("-", "/", regex=False)
)

df.info()
```

```python
df["order_date"] = pd.to_datetime(
    df["order_date"],
    errors="coerce",
    dayfirst=False
)

df.info()
```

```python
engine = create_engine(
    "mysql+pymysql://root:database123@localhost:3306/balaji_food"
)
```

```python
with engine.connect() as conn:
    print("Koneksi berhasil!")
```

```python
df.to_sql(
    name="sales",
    con=engine,
    if_exists="append",
    index=False
)
```
